```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Data/laptopData.csv")
```

```python
df = df.dropna(subset=['Company'])
```

```python
sns.histplot(data=df, x=df['Price'])
```

```python
# Absolute Maximum Scaling

# Formula: (X-max)/max
```

```python
# MinMax Scaler
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, Normalizer
```

```python
mms = MinMaxScaler()
scaled_data = mms.fit_transform(df[['Price']])
d = pd.DataFrame(scaled_data, columns=['Price'])
```

```python
sns.histplot(data=d, x=d['Price'])
```

```python
#Normalization

# Formula = (X-mean) / Range 
```

```python
df["Price"].head()
```

```python
norm = Normalizer()
nd = norm.fit_transform(df[['Price']])
dd = pd.DataFrame(nd, columns=["price"])
```

```python
sns.histplot(data=dd, x=dd['price'])
```

```python
#StandardScaler

# Formula: (X-mean) / standard deviation
```

```python
ss = StandardScaler()
sc_dt = ss.fit_transform(df[['Price']])
d1 = pd.DataFrame(sc_dt, columns=['Price'])
```

```python
sns.histplot(data=d1, x= d1['Price'])
```

```python
# Robust Scaling

# (X - median) / IQR 
```

```python
 from sklearn.preprocessing import RobustScaler
```

```python
rs = RobustScaler()
scd = rs.fit_transform(df[['Price']])
d2 = pd.DataFrame(scd, columns=["Price"])
```

```python
sns.histplot(data=d2, x=d2['Price'])
```
