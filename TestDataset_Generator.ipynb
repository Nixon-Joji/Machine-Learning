```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs, make_circles, make_moons, make_classification, make_multilabel_classification
```

```python
X, y = make_circles(n_samples=1000, noise=0.1, random_state=8, shuffle=True)

plt.scatter(X[:,0], X[:,1], s=20, c=y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
X, y = make_moons(n_samples=1000, noise=0.1, random_state=8, shuffle=True)

plt.scatter(X[:,0], X[:,1], s=10, c=y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
X,y = make_blobs(n_samples=100, n_features=2, centers=3, random_state=23)
plt.scatter(X[:,0], X[:,1], s=20, c=y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

MAKE CLASSIFICATION

```python
X,y = make_classification(n_classes=2,
                          n_features=10,
                          n_samples=1000,
                          n_redundant=2,
                          n_informative=7,
                          n_repeated=1,
                          n_clusters_per_class=1,
                          random_state=42)

plt.scatter(X[:,0], X[:,1], s=20, c=y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
X,y = make_multilabel_classification(n_classes=2,
                               n_labels=3,
                               n_samples=100,
                               n_features=2,
                               allow_unlabeled=True,
                               random_state=3
                               )
```

```python
df = pd.concat([pd.DataFrame(X, columns=['X1', 'X2']),
                pd.DataFrame(y, columns=['Label1', 'Label2'])],
                axis=1)

df.head()
```

```python
plt.scatter(df['X1'], df['X2'], c=df['Label1'])
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
from sklearn.datasets import make_regression
```

```python
from sklearn.datasets import make_sparse_uncorrelated
```

```python
X,y = make_sparse_uncorrelated(n_samples=100, n_features=5, random_state=8)
#Why minimum 4?
```

```python
for i in range(5):
    plt.subplot(2,3, i+1)
    plt.scatter(X[:,i], y)
    plt.xlabel("X"+str(i))
    plt.ylabel("Y")
plt.show()
```

```python
from sklearn.datasets import make_friedman2
```

```python
X,y = make_friedman2(n_samples=100, random_state=8)
for i in range(4):
    plt.subplot(2,2, i+1)
    plt.scatter(X[:,i], y)
    plt.xlabel('X'+str(i+1))
    plt.ylabel('Y')
plt.show()
```
