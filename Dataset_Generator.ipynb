```python
from sklearn.datasets import make_blobs, make_circles, make_moons
from matplotlib import pyplot as plt
```

```python
X, y = make_blobs(n_samples=100,
                  centers=3,
                  cluster_std=1,
                  n_features=2)

plt.scatter(X[:,0], X[:,1], s=20, color = 'g')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
X, y = make_moons(n_samples=1000, noise=0.1)

plt.scatter(X[:,0], X[:,1], s=20, color = 'g')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
