```python
from sklearn.datasets import make_blobs
from matplotlib import pyplot as plt, style 
```

```python
style.use("fivethirtyeight")
```

```python
X, y = make_blobs(n_samples=100,
                  centers=3,
                  cluster_std=5,
                  n_features=9)
```

```python
plt.scatter(X[:,0], X[:,1], s=40, color = 'g')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
# plt.clf()
```
